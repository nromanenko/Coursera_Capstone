# Getting the latitude and the longitude coordinates of each neighborhood in Toronto#

This notebook is created for part II of 'Segmenting and Clustering Neighborhoods in Toronto' Assignment. Here we use the dataframe from previous part and get the latitude and the longitude coordinates of each neighborhood.

````python
import pandas as pd
````

````python
# read csv with Toronto neighborhood data
df = pd.read_csv('neighborhoods.csv')
df.head(10)
````

So I decided to get the data from the provided csv

````python
geo_df = pd.read_csv("Geospatial_Coordinates.csv")
geo_df.head(10)
````

````python
# let's join both dataframes on Postal Code
df = df.join(geo_df.set_index('Postal Code'), on='PostalCode')
df.head(10)
````

````python
# save to csv file to use later
df.to_csv('neighborhoods_geo.csv', index=False)
````
